##Import

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
from tqdm.notebook import tqdm
```

##Pull Data

The code below will scrape for tweets by a text search then provide a CSV file with Pandas

```python
# Initialize an empty list to store the tweets
tweets = []

# Set the limit of tweets to scrape
limit = 50

# Define the words to search for in tweets
words = 'ftx AND binance' # does not matter if the query is lower or upper case

# Define the date range (inclusive) for scraping tweets
start_date = "2022-09-01"
end_date = "2023-01-12"

# Define the language to search for
language = "en"

# Create a query string using the words and date range
query = words + " since:" + start_date + " until:" + end_date + " lang:" + language

# Using TwitterSearchScraper to scrape data and append tweets to list
for tweet in tqdm(sntwitter.TwitterSearchScraper(query).get_items(), total = limit):

        if len(tweets) == limit:
         break
        else:
             tweets.append([tweet.user.username, tweet.user.location, tweet.user.followersCount, tweet.id, tweet.likeCount, tweet.quoteCount, tweet.replyCount, tweet.retweetCount, tweet.date, tweet.content])
```

```python
# Columns
cols = ['username', 'user_location', 'follower_count', 'id', 'like_count', 'quote_count', 'reply_count', 'retweet_count', 'date', 'tweet']

# Create a dataframe from the tweets list above
df = pd.DataFrame(tweets, columns = cols)

# Display first 5 entries from dataframe
df.head(10)
```

```python
# Display last 5 entries from dataframe
df.tail()
```

## Export Data

```python
# Export dataframe into a CSV
df.to_csv('ftx-binance-tweets.csv', index=False, sep=',', header=True)
```
